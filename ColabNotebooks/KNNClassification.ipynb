Problem Statement : Assignment on KNN Classification
In the following diagram let blue circles indicate positive examples and orange squares indicate negative examples.
We want to use k-NN algorithm for classifying the points. If k=3, find the class of the point (6,6).
Extend the same example for Distance-Weighted k-NN and Locally weighted Averaging.


```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
```

```python
dataset = pd.DataFrame({'x':[2,4,4,4,6,6],
                        'y':[4,2,4,6,2,4],
                        'class':['negative','negative','positive','negative','positive','negative']})
dataset
```

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
x
```

```python
y
```

# General KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x,y)
knn_y_pred = knn.predict(x) 
knn_y_pred
```

```python
print(str(accuracy_score(y,knn_y_pred)))
print(str(confusion_matrix(y,knn_y_pred)))
print(str(f1_score(y,knn_y_pred,average=None)))
```

# Distance weighted KNN

```python
w_knn = KNeighborsClassifier(n_neighbors =3, weights = 'distance')
w_knn.fit(x,y)
w_knn_y_pred = w_knn.predict(x) 
w_knn_y_pred
```

```python
print(str(accuracy_score(y,w_knn_y_pred)))
print(str(confusion_matrix(y,w_knn_y_pred)))
print(str(f1_score(y,w_knn_y_pred,average=None)))
```

# Local weighted averaging

```python
from sklearn.neighbors import NearestCentroid
LWA = NearestCentroid()
LWA.fit(x,y)
print(LWA.centroids_)
LWA_y_pred = LWA.predict(x)
LWA_y_pred
```

```python
print(str(accuracy_score(y,LWA_y_pred)))
print(str(confusion_matrix(y,LWA_y_pred)))
print(str(f1_score(y,LWA_y_pred,average=None)))
```

```python
print('Class of the point (6,6) : ')
print('by General K-NN : ',''.join(knn.predict([np.array([6,6])])))
print('by Weighted K-NN : ',''.join(w_knn.predict([np.array([6,6])])))
print('by Locally Weighted Averaging : ',''.join(LWA.predict([np.array([6,6])])))
```
