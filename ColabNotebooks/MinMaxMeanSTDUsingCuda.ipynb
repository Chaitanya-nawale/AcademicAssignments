```
%%cu
#include <stdio.h>
#include <stdlib.h>
#include<time.h>
#include <cuda.h>

#define SIZE 512

__global__ void minFind(int *a, int *c){
    int of = SIZE/2 ;
    int i = threadIdx.x ;
    for(of;of>0;of=of/2){
        if(i<of){
            if(a[i]>a[i+of]){
                a[i] = a[i+of] ;
            }
        }
    __syncthreads();
    }
    c[0] = a[0] ;
}
int main(){
    int i ;
    srand(time(NULL)) ;
    int *a = (int*)malloc(SIZE*sizeof(int)) ;
    int *c = (int*)malloc(SIZE*sizeof(int)) ;
    int *dev_a,*dev_c ;
    cudaMalloc((void**)&dev_a,SIZE*sizeof(int)) ;
    cudaMalloc((void**)&dev_c,SIZE*sizeof(int)) ;
    for(i=0;i<SIZE;i++){
        a[i] = rand() ;
    }
    for(i=0;i<SIZE;i++){
        printf("%d ",a[i]) ;
    }
    printf("\n") ;
    cudaMemcpy(dev_a,a,SIZE*sizeof(int),cudaMemcpyHostToDevice) ;
    minFind<<<1,SIZE/2>>>(dev_a,dev_c) ;
    cudaMemcpy(c,dev_c,SIZE*sizeof(int),cudaMemcpyDeviceToHost) ;
    printf("%d",c[0]) ;
    cudaFree(dev_a) ;
    cudaFree(dev_c) ;
    free(a) ;
    free(c) ;
    return 0 ;
}
```

```
%%cu
#include <stdio.h>
#include <stdlib.h>
#include<time.h>
#include <cuda.h>

#define SIZE 512

__global__ void sumOf(int *a, int *c){
    int of = SIZE/2 ;
    int i = threadIdx.x ;
    for(of;of>0;of=of/2){
        if(i<of){
            a[i] += a[i+of] ;
        }
        __syncthreads();
    }
    c[0] = a[0] ;
}

__global__ void devFind(int *a,int avg){
    int tid = threadIdx.x ;
    if(tid<SIZE){
        a[tid] -= avg ;
        a[tid] = a[tid] * a[tid] ;
    }    
}

int main(){
    int i ;
    srand(time(NULL)) ;
    int *a = (int*)malloc(SIZE*sizeof(int)) ;
    int *c = (int*)malloc(SIZE*sizeof(int)) ;
    int *d = (int*)malloc(SIZE*sizeof(int)) ;
    int *dev_a,*dev_c,*dev_d,*dev_f ;
    cudaMalloc((void**)&dev_a,SIZE*sizeof(int)) ;
    cudaMalloc((void**)&dev_c,SIZE*sizeof(int)) ;
    cudaMalloc((void**)&dev_d,SIZE*sizeof(int)) ;
    cudaMalloc((void**)&dev_f,SIZE*sizeof(int)) ;
    for(i=0;i<SIZE;i++){
        a[i] = i+1 ;
    }
    for(i=0;i<SIZE;i++){
        printf("%d ",a[i]) ;
    }
    printf("\n") ;
    cudaMemcpy(dev_a,a,SIZE*sizeof(int),cudaMemcpyHostToDevice) ;
    cudaMemcpy(dev_d,a,SIZE*sizeof(int),cudaMemcpyHostToDevice) ;
    sumOf<<<1,SIZE/2>>>(dev_a,dev_c) ;
    cudaMemcpy(c,dev_c,SIZE*sizeof(int),cudaMemcpyDeviceToHost) ;
    float op = (float) c[0] / SIZE;
    devFind<<<1,SIZE>>>(dev_d,op) ;
    sumOf<<<1,SIZE/2>>>(dev_d,dev_f) ;
    cudaMemcpy(d,dev_f,SIZE*sizeof(int),cudaMemcpyDeviceToHost) ;
    int temp = d[0] / SIZE ;
    printf("%f",sqrt(temp)) ;
    cudaFree(dev_a) ;
    cudaFree(dev_c) ;
    cudaFree(dev_d) ;
    cudaFree(dev_f) ;
    free(a) ;
    free(c) ;
    return 0 ;
}
```
