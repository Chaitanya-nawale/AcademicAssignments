```
import string
alphabet_string = string.ascii_uppercase
alphabet_list = list(alphabet_string)
print(alphabet_list)
```

```
for i in alphabet_string:
  if i  not in l:
    if i!="J":
      l.append(i)
    else:
      if "I" in l:
        pass

      else:
        l.append("l")

x=0
key_matrix= [["-" for _ in range(5)]for _ in range(5)]
for i in range(0,5): #making matrix
    for j in range(0,5):
        key_matrix[i][j]=l[x]
        x+=1

print("\nkey = "+ key+"\n")
for i in key_matrix:
  print(i)
```

#**2) Take input text and apply rules**

```
def get_location(ch):
  if ch=="J":
    ch="I"
  for i in range(5):
    for j in range(5):
      if(key_matrix[i][j] == ch):
        return [i,j]
```

```
def encrypt_message(ip):
  op=""
  x=0
  while(x<len(ip)):
    loc1 = get_location(ip[x])
    loc2 = get_location(ip[x+1])

    #same rows => immediate right
    if loc1[0]==loc2[0]:
      l1 = (loc1[1]+1)%5
      l2 = (loc2[1]+1)%5
      op+=key_matrix[loc1[0]][l1]   #row same col+1
      op+=key_matrix[loc2[0]][l2]
      op+=" "

    #same column =>below
    elif loc1[1]==loc2[1]:
      l1 = (loc1[0]+1)%5
      l2 = (loc2[0]+1)%5
      op+=key_matrix[l1][loc1[1]]  #col same , row+1
      op+=key_matrix[l2][loc2[1]]
      op+=" "

    else:
      #opp diagonal
      op+=key_matrix[loc1[0]][loc2[1]]  
      op+=key_matrix[loc2[0]][loc1[1]]
      op+=" "  


    
    x+=2
  return op
```

```
def decrypt_message(ip):
  op=""
  x=0
  while(x<len(ip)):
    loc1 = get_location(ip[x])
    loc2 = get_location(ip[x+1])
    

    #same rows => immediate left
    if loc1[0]==loc2[0]:
      l1 = (loc1[1]+4)%5
      l2 = (loc2[1]+4)%5
      op+=key_matrix[loc1[0]][l1]   
      op+=key_matrix[loc2[0]][l2]
      op+=" "

    #same column =>above
    elif loc1[1]==loc2[1]:
      l1 = (loc1[0]+4)%5
      l2 = (loc2[0]+4)%5
      op+=key_matrix[l1][loc1[1]]  #col same , row+1
      op+=key_matrix[l2][loc2[1]]
      op+=" "

    else:
      #opp diagonal
      op+=key_matrix[loc1[0]][loc2[1]]  
      op+=key_matrix[loc2[0]][loc1[1]]
      op+=" "  


    
    x+=2
  return op
```

```
ip = input()
ip = ip.replace(" ","")
ip = ip.upper()
ip = [i for i in ip]

if(len(ip)%2==1):
  ip.append("X")
op = encrypt_message(ip)
print("Cipher text")
print(op)
ip2 = op.replace(" ","")
ip2= [i for i in ip2]
op1 = decrypt_message(ip2)
print("\n plain text")
print(op1)
```
